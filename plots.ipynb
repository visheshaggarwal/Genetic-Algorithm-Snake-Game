```python
import matplotlib.pyplot as plt
```

```python
basicBarrierMax = []
basicBarrierAvg = []
for i in range(80):
    f = open("basicBarrier/uniformCrossover/gen_"+str(i)+"_best", 'r')
    c = f.read()
    c = c.split(',')
    basicBarrierMax.append(float(c[-2]))
    basicBarrierAvg.append(float(c[-1]))

print(basicBarrierAvg,basicBarrierMax)
```

```python
basicFood_basicBarrierMax = []
basicFood_basicBarrierAvg = []
for i in range(80):
    f = open("basicFood-basicBarrier/uniformCrossover/gen_"+str(i)+"_best", 'r')
    c = f.read()
    c = c.split(',')
    basicFood_basicBarrierMax.append(float(c[-2]))
    basicFood_basicBarrierAvg.append(float(c[-1]))
```

```python
basicFood_scaledBarrierMax = []
basicFood_scaledBarrierAvg = []
for i in range(80):
    f = open("basicFood-scaledBarrier/uniformCrossover/gen_"+str(i)+"_best", 'r')
    c = f.read()
    c = c.split(',')
    basicFood_scaledBarrierMax.append(float(c[-2]))
    basicFood_scaledBarrierAvg.append(float(c[-1]))
```

```python
basicFood_scaledBarrier_rewardMax = []
basicFood_scaledBarrier_rewardAvg = []
for i in range(80):
    f = open("basicFood-scaledBarrier-reward/uniformCrossover/gen_"+str(i)+"_best", 'r')
    c = f.read()
    c = c.split(',')
    basicFood_scaledBarrier_rewardMax.append(float(c[-2]))
    basicFood_scaledBarrier_rewardAvg.append(float(c[-1]))
```

```python
y = []
for i in range(80):
    y.append(i)
```

```python
plt.figure(figsize=(10,5))
plt.plot(y,basicBarrierMax,label="basicBarrier")
plt.plot(y,basicFood_scaledBarrier_rewardMax,label="basicFood-scaledBarrier-reward")
plt.plot(y,basicFood_scaledBarrierMax,label="basicFood-scaledBarrier")
plt.plot(y,basicFood_basicBarrierMax,label="basicFood-basicBarrier")
plt.xlabel('Generation')
plt.ylabel('Max Score')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(y,basicBarrierAvg,label="basicBarrier")
plt.plot(y,basicFood_scaledBarrier_rewardAvg,label="basicFood-scaledBarrier-reward")
plt.plot(y,basicFood_scaledBarrierAvg,label="basicFood-scaledBarrier")
plt.plot(y,basicFood_basicBarrierAvg,label="basicFood-basicBarrier")
plt.xlabel('Generation')
plt.ylabel('Avg Score')
plt.legend()
plt.show()
```
